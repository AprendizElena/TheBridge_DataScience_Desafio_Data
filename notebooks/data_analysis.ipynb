```python
import pandas as pd
from ydata_profiling import ProfileReport
import os
import glob
import webbrowser
```

```python
def load_most_recent_csv(folder_path):
    # Use glob to find all CSV files in the specified folder
    csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
    
    # Get the most recent file based on the modification time
    if csv_files:
        latest_file = max(csv_files, key=os.path.getmtime)
        print(f"Loading the most recent file: {latest_file}")
        return pd.read_csv(latest_file)
    else:
        print("No CSV files found in the specified folder.")
        return None
```

```python
folder_path = '../data'
data = load_most_recent_csv(folder_path)
```

```python
print(data.head())
```

```python
profile = ProfileReport(data, title='Pandas Profiling Report', explorative=True)

report_folder = "../data_report"  
os.makedirs(report_folder, exist_ok=True)  
report_filename = os.path.join(report_folder, "data_report.html")

profile.to_file(report_filename)

webbrowser.open('file://' + os.path.realpath(report_filename))

profile.to_notebook_iframe()
```
